```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 1. Create a directed graph object
G = nx.DiGraph()

# 2. Add nodes (optional, adding edges will automatically add nodes)
G.add_nodes_from(["A", "B", "C", "D"])

# 3. Add directed edges
G.add_edges_from([("A", "B"), ("B", "C"), ("C", "A"), ("A", "D")])

# 4. Define the layout for the nodes
# This helps in positioning the nodes for better visualization.
# Different layouts exist, e.g., spring_layout, circular_layout, shell_layout.
pos = nx.spring_layout(G) 

# 5. Draw the graph
# arrows=True is crucial for directed graphs to show the direction of edges.
nx.draw_networkx(G, pos, with_labels=True, arrows=True, node_color='skyblue', node_size=1000, font_size=10)

# 6. Display the plot
plt.title("Simple Directed Graph")
plt.axis('off') # Turn off the axis
plt.show()
```

```python
G = nx.DiGraph()
G.add_nodes_from(["A", "B", "C", "D"])
G.add_edges_from([("A", "B"), ("B", "C"), ("C", "D"), ("D", "A")])
pos = nx.spring_layout(G) 
nx.draw_networkx(G, pos, with_labels=True, arrows=True, node_color='skyblue', node_size=1000, font_size=10)
plt.title("Cyclic Graph")
plt.axis('off')
plt.show()
```

```python
G = nx.DiGraph()
edges = [
    (1, 2), (2, 3), (3, 1),  # SCC {1,2,3}
    (3, 4),                  # edge out of SCC
    (4, 5), (5, 6), (6, 4),  # SCC {4,5,6}
    (6, 7)                   # tail
]
G.add_edges_from(edges)

# Find strongly connected components
sccs = list(nx.strongly_connected_components(G))
print("Strongly connected components:", sccs)

# Assign a color per SCC
color_map = {}
for i, comp in enumerate(sccs):
    for node in comp:
        color_map[node] = i

node_colors = [color_map[node] for node in G.nodes()]

# Draw the graph
pos = nx.spring_layout(G, seed=42)  # layout for better visualization
plt.figure(figsize=(8, 6))
nx.draw(
    G, pos,
    with_labels=True,
    node_color=node_colors,
    node_size=1000,
    cmap=plt.cm.Set3,  # distinct colors
    arrowsize=20,
    font_size=12
)

plt.title("Directed Graph with Strongly Connected Components")
plt.show()
```

```python

# Create a complete bipartite graph with 3 nodes in set A and 2 nodes in set B
G = nx.complete_bipartite_graph(3, 2)

# Separate nodes by bipartite sets
A = {0, 1, 2}   # first partition
B = {3, 4}      # second partition

# Position nodes: bipartite_layout requires specifying one partition
pos = nx.bipartite_layout(G, A)

# Draw graph
plt.figure(figsize=(6, 6))
nx.draw(
    G, pos,
    with_labels=True,
    node_color=["skyblue" if n in A else "lightgreen" for n in G.nodes()],
    node_size=1000,
    edge_color="gray",
    width=2,
    font_size=12
)

plt.title("Complete Bipartite Graph K(3,2)")
plt.show()
```

```python
# Create a complete graph with 5 nodes
G = nx.complete_graph(5)

# Choose layout (circular looks nice for complete graphs)
pos = nx.circular_layout(G)

# Draw the graph
plt.figure(figsize=(6, 6))
nx.draw(
    G, pos,
    with_labels=True,
    node_color="skyblue",
    node_size=1000,
    edge_color="gray",
    width=2,
    font_size=12
)

plt.title("Complete Graph K5")
plt.show()
```

```python
G = nx.complete_graph(4)
pos = nx.circular_layout(G)
plt.figure(figsize=(5, 5))
nx.draw(
    G, pos,
    with_labels=True,
    node_color="skyblue",
    node_size=500,
    edge_color="gray",
    width=2,
    font_size=8
)

plt.title("Complete Graph K4")
plt.show()
```

```python

# Create complete graph with 4 nodes
G = nx.complete_graph(4)

# Define manual positions (square layout)
pos = {
    0: (0, 0),
    1: (1, 0),
    2: (1, 1),
    3: (0, 1)
}

plt.figure(figsize=(4, 4))
nx.draw(
    G, pos,
    with_labels=True,
    node_color="skyblue",
    node_size=800,
    edge_color="gray",
    width=2,
    font_size=8
)

plt.title("Planar Drawing of K4 (no edge crossings)")
plt.show()
```
