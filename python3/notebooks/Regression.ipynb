```python
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
```

```python
X= np.array([0,1,2,3,4,5]).reshape(-1, 1)
```

```python
y=np.array([0,4,8,12,16,20]).reshape(-1, 1) #label
```

```python
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
actaul = np.array([6,7]).reshape(-1, 1)
```

```python

prediction = model.predict(actaul)
```

```python
prediction
```

```python
print("Slope (d):", model.coef_[0])
print("Intercept:", model.intercept_)
```
