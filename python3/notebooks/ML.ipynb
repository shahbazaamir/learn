```python
import numpy as np

# Generate AP data
n = 20
a = 2      # first term
d = 3      # common difference

X = np.arange(1, n + 1).reshape(-1, 1)   # n values
y = a + (X - 1) * d                      # arithmetic progression
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)

print("Slope (d):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.plot(X, y_pred)
plt.xlabel("n")
plt.ylabel("Value")
plt.show()
```
